```
using System;
using System.Collections.Generic;

public class CreditCard
{
    public string CardNumber { get; set; }
    public string HolderName { get; set; }
    public string ExpiryDate { get; set; }
    protected decimal Balance { get; set; }

    public CreditCard(string cardNumber, string holderName, string expiryDate, decimal balance = 0)
    {
        CardNumber = cardNumber;
        HolderName = holderName;
        ExpiryDate = expiryDate;
        Balance = balance;
    }

    public virtual string GetInfo()
    {
        return $"Карта: {CardNumber}\nВладелец: {HolderName}\nСрок действия: {ExpiryDate}";
    }

    public virtual string Pay(decimal amount)
    {
        if (Balance >= amount)
        {
            Balance -= amount;
            return $"Оплата на сумму {amount:C} прошла успешно!";
        }
        else
        {
            return $"Недостаточно средств! Баланс: {Balance:C}";
        }
    }

    public virtual string CheckBalance()
    {
        return $"Баланс карты: {Balance:C}";
    }
}

public class GoldCreditCard : CreditCard
{
    public int BonusMiles { get; set; }

    public GoldCreditCard(string cardNumber, string holderName, string expiryDate, decimal balance = 0, int bonusMiles = 0)
        : base(cardNumber, holderName, expiryDate, balance)
    {
        BonusMiles = bonusMiles;
    }

    public override string Pay(decimal amount)
    {
        if (Balance >= amount)
        {
            Balance -= amount;
            //(1 миля за каждые 100руб)
            int earnedMiles = (int)(amount / 100);
            BonusMiles += earnedMiles;
            return $"Оплата на сумму {amount:C} прошла успешно! Начислено {earnedMiles} бонусных миль.";
        }
        else
        {
            return $"Недостаточно средств! Баланс: {Balance:C}";
        }
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $"\nТип: Gold\nБонусные мили: {BonusMiles}";
    }
}

public class PremiumCreditCard : CreditCard
{
    public string SupportAssistant { get; set; }

    public PremiumCreditCard(string cardNumber, string holderName, string expiryDate, string supportAssistant, decimal balance = 0)
        : base(cardNumber, holderName, expiryDate, balance)
    {
        SupportAssistant = supportAssistant;
    }

    public override string CheckBalance()
    {
        if (Balance < 1000)
        {
            return $"Баланс карты: {Balance:C}\nВнимание: низкий баланс! Обратитесь к ассистенту поддержки: {SupportAssistant}";
        }
        return base.CheckBalance();
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $"\nТип: Premium\nАссистент поддержки: {SupportAssistant}";
    }
}

public class CorporateCreditCard : CreditCard
{
    public string Company { get; set; }

    public CorporateCreditCard(string cardNumber, string holderName, string expiryDate, string company, decimal balance = 0)
        : base(cardNumber, holderName, expiryDate, balance)
    {
        Company = company;
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $"\nТип: Corporate\nКомпания: {Company}";
    }
}

public class BankAccount
{
    public string AccountNumber { get; set; }
    public string OwnerName { get; set; }
    protected decimal Balance { get; set; }

    public BankAccount(string accountNumber, string ownerName, decimal balance = 0)
    {
        AccountNumber = accountNumber;
        OwnerName = ownerName;
        Balance = balance;
    }

    public virtual void Deposit(decimal amount)
    {
        Balance += amount;
        Console.WriteLine($"Пополнение на {amount:C}. Новый баланс: {Balance:C}");
    }

    public virtual bool Withdraw(decimal amount)
    {
        if (Balance >= amount)
        {
            Balance -= amount;
            Console.WriteLine($"Снятие {amount:C}. Остаток: {Balance:C}");
            return true;
        }
        Console.WriteLine($"Недостаточно средств! Баланс: {Balance:C}");
        return false;
    }

    public virtual string GetAccountInfo()
    {
        return $"Счет: {AccountNumber}\nВладелец: {OwnerName}\nБаланс: {Balance:C}";
    }
}

public class SavingsAccount : BankAccount
{
    public decimal InterestRate { get; set; }

    public SavingsAccount(string accountNumber, string ownerName, decimal interestRate, decimal balance = 0)
        : base(accountNumber, ownerName, balance)
    {
        InterestRate = interestRate;
    }

    public void ApplyInterest()
    {
        decimal interest = Balance * InterestRate / 100;
        Balance += interest;
        Console.WriteLine($"Начислены проценты: {interest:C}. Новый баланс: {Balance:C}");
    }

    public override string GetAccountInfo()
    {
        return base.GetAccountInfo() + $"\nТип: Сберегательный\nПроцентная ставка: {InterestRate}%";
    }
}

public class InvestmentAccount : SavingsAccount
{
    public string InvestmentType { get; set; }
    public decimal RiskLevel { get; set; }

    public InvestmentAccount(string accountNumber, string ownerName, decimal interestRate, string investmentType, decimal riskLevel, decimal balance = 0)
        : base(accountNumber, ownerName, interestRate, balance)
    {
        InvestmentType = investmentType;
        RiskLevel = riskLevel;
    }

    public void SimulateInvestmentGrowth(int years)
    {
        decimal futureBalance = Balance;
        for (int i = 0; i < years; i++)
        {
            futureBalance *= (1 + InterestRate / 100);
        }
        Console.WriteLine($"Через {years} лет баланс составит: {futureBalance:C}");
    }

    public override string GetAccountInfo()
    {
        return base.GetAccountInfo() + $"\nТип: Инвестиционный\nВид инвестиций: {InvestmentType}\nУровень риска: {RiskLevel}";
    }
}

public interface ITransferable
{
    bool TransferTo(BankAccount targetAccount, decimal amount);
    string GetTransferInfo();
}

public interface IBlockable
{
    void BlockAccount();
    void UnblockAccount();
    bool IsBlocked { get; }
}

public class CurrentAccount : BankAccount, ITransferable, IBlockable
{
    public decimal OverdraftLimit { get; set; }
    public bool IsBlocked { get; private set; }

    public CurrentAccount(string accountNumber, string ownerName, decimal overdraftLimit, decimal balance = 0)
        : base(accountNumber, ownerName, balance)
    {
        OverdraftLimit = overdraftLimit;
        IsBlocked = false;
    }

    public override bool Withdraw(decimal amount)
    {
        if (IsBlocked)
        {
            Console.WriteLine("Счет заблокирован! Операция невозможна.");
            return false;
        }

        if (Balance + OverdraftLimit >= amount)
        {
            Balance -= amount;
            Console.WriteLine($"Снятие {amount:C}. Остаток: {Balance:C}");
            return true;
        }
        Console.WriteLine($"Превышен лимит овердрафта! Баланс: {Balance:C}, Лимит: {OverdraftLimit:C}");
        return false;
    }

    public bool TransferTo(BankAccount targetAccount, decimal amount)
    {
        if (IsBlocked)
        {
            Console.WriteLine("Счет заблокирован! Перевод невозможен.");
            return false;
        }

        if (Withdraw(amount))
        {
            targetAccount.Deposit(amount);
            Console.WriteLine($"Перевод {amount:C} на счет {targetAccount.AccountNumber} выполнен успешно");
            return true;
        }
        return false;
    }

    public string GetTransferInfo()
    {
        return $"Текущий счет для переводов: {AccountNumber}";
    }

    public void BlockAccount()
    {
        IsBlocked = true;
        Console.WriteLine($"Счет {AccountNumber} заблокирован");
    }

    public void UnblockAccount()
    {
        IsBlocked = false;
        Console.WriteLine($"Счет {AccountNumber} разблокирован");
    }

    public override string GetAccountInfo()
    {
        string status = IsBlocked ? "ЗАБЛОКИРОВАН" : "АКТИВЕН";
        return base.GetAccountInfo() + $"\nТип: Текущий\nЛимит овердрафта: {OverdraftLimit:C}\nСтатус: {status}";
    }
}

        Console.WriteLine("=== Демонстрация работы с кредитными картами ===\n");
        
        CreditCard basicCard = new CreditCard("1234 5678 9012 3456", "Иван Иванов", "12/25", 5000);
        GoldCreditCard goldCard = new GoldCreditCard("2345 6789 0123 4567", "Петр Петров", "06/26", 10000, 150);
        PremiumCreditCard premiumCard = new PremiumCreditCard("3456 7890 1234 5678", "Мария Сидорова", "09/27", "support@premiumbank.ru", 800);
        CorporateCreditCard corporateCard = new CorporateCreditCard("4567 8901 2345 6789", "Алексей Козлов", "03/28", "ООО 'Технологии'", 20000);

        Console.WriteLine("=== Базовая карта ===");
        Console.WriteLine(basicCard.GetInfo());
        Console.WriteLine(basicCard.CheckBalance());
        Console.WriteLine(basicCard.Pay(3000));
        Console.WriteLine(basicCard.CheckBalance());
        Console.WriteLine();

        Console.WriteLine("=== Золотая карта ===");
        Console.WriteLine(goldCard.GetInfo());
        Console.WriteLine(goldCard.CheckBalance());
        Console.WriteLine(goldCard.Pay(2500));
        Console.WriteLine(goldCard.GetInfo());
        Console.WriteLine();

        Console.WriteLine("=== Премиум карта ===");
        Console.WriteLine(premiumCard.GetInfo());
        Console.WriteLine(premiumCard.CheckBalance());
        Console.WriteLine(premiumCard.Pay(500));
        Console.WriteLine(premiumCard.CheckBalance());
        Console.WriteLine();

        Console.WriteLine("=== Корпоративная карта ===");
        Console.WriteLine(corporateCard.GetInfo());
        Console.WriteLine(corporateCard.CheckBalance());
        Console.WriteLine(corporateCard.Pay(15000));
        Console.WriteLine(corporateCard.CheckBalance());
        Console.WriteLine();

        Console.WriteLine("=== Полиморфизм ===");
        CreditCard[] cards = { basicCard, goldCard, premiumCard, corporateCard };
        
        foreach (var card in cards)
        {
            Console.WriteLine(card.GetInfo());
            Console.WriteLine(card.Pay(1000));
            Console.WriteLine(card.CheckBalance());
            Console.WriteLine("---");
        }

        Console.WriteLine("\n=== ДОПОЛНИТЕЛЬНОЕ ЗАДАНИЕ: Типы наследования ===\n");

        Console.WriteLine("=== Простое наследование ===");
        SavingsAccount savings = new SavingsAccount("SAV001", "Анна Петрова", 5.0m, 50000);
        Console.WriteLine(savings.GetAccountInfo());
        savings.ApplyInterest();
        Console.WriteLine();

        Console.WriteLine("=== Сложное наследование (цепочка) ===");
        InvestmentAccount investment = new InvestmentAccount("INV001", "Сергей Смирнов", 8.5m, "Акции", 0.7m, 100000);
        Console.WriteLine(investment.GetAccountInfo());
        investment.SimulateInvestmentGrowth(5);
        Console.WriteLine();

        Console.WriteLine("=== Множественное наследование (интерфейсы) ===");
        CurrentAccount current = new CurrentAccount("CUR001", "Дмитрий Орлов", 5000, 2000);
        Console.WriteLine(current.GetAccountInfo());
        

        Console.WriteLine(current.GetTransferInfo());
        current.TransferTo(savings, 1000);
        current.BlockAccount();
        current.TransferTo(savings, 500);
        current.UnblockAccount();
        Console.WriteLine(current.GetAccountInfo());
```
